```python
import pandas as pd
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/codebasics/py/master/ML/9_decision_tree/salaries.csv")
```

```python
df
```

```python
input = df.drop(['salary_more_then_100k'],axis=1)
```

```python
input.head()
```

```python
target=df.salary_more_then_100k
```

```python
target.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le_company = LabelEncoder()
le_job = LabelEncoder()
le_degree = LabelEncoder()
```

```python
input['company_n']=le_company.fit_transform(input['company'])
input['job_n']=le_job.fit_transform(input['job'])
input['degree_n']=le_degree.fit_transform(input['degree'])
```

```python
input
```

```python
inputs_n = input.drop(['company','job','degree'],axis='columns')
```

```python
inputs_n.head()
```

```python
from sklearn import tree
```

```python
model = tree.DecisionTreeClassifier()
```

```python
model.fit(inputs_n,target)
```

```python
model.score(inputs_n,target)
```

```python
model.predict([[2,0,0]])
```
